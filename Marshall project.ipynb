```python
import pandas as pd
crime = pd.read_csv('Downloads/crime.csv')
crime=crime.dropna()
```

```python
crime.groupby('report_year')['crimes_percapita'].mean().plot()
```

```python
import numpy as np

bins = [0, 125000, 250000, 375000, 500000, 625000, 750000, np.inf]
names = ['0-125k', '125k-250k','250k-375k', '375k-500k', '500k-625k','625k-750k','750k+']

crime['CitySize'] = pd.cut(crime['population'], bins, labels=names)

crime.groupby('CitySize')['crimes_percapita'].mean().plot()
```
